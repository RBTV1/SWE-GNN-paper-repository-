# Simulation generation

### DEM generator 

Change seed, noise_octave, or DEM_multiplier to generate different topographies

```python
import matplotlib.pyplot as plt
import numpy as np 

from dhydro_utils import grid_DEM_generator

seed = 1
DEM = grid_DEM_generator(seed, number_grids=64, noise_octave=5, DEM_multiplier=6)
plt.imshow(DEM, cmap='terrain', origin='lower')
plt.colorbar();
```

### Run simulations

Run D-HYDRO numerical model (requires DELFT3D-FM (or D-HYDRO) license and domain setup)

```python
from dhydro_utils import run_simulations

number_grids = 64

model_folder = 'C:\\Users\\rbentivoglio\\Documents\\PhD\\My PhD\\D-HYDRO\\SWEGNN_128.dsproj_data\\SWE_GNN_dataset'
save_folder = 'raw_datasets'

df = run_simulations(10, model_folder, save_folder, start_sim=16001, number_grids=number_grids, noise_octave=8, initial_octave=14, DEM_multiplier=5, grid_size=100, random_breach=True)
```

### Read results

```python
sim = 12

fig, axs = plt.subplots(1,2,figsize=(14,6))
axs[0].imshow(np.loadtxt(f"{save_folder}\\DEM\\DEM_{sim}.txt")[:, 2].reshape(number_grids,number_grids), cmap='terrain', origin='lower');
plt.imshow(np.loadtxt(f"{save_folder}\\WD\\WD_{sim}.txt")[-1].reshape(number_grids,number_grids), cmap='Blues', origin='lower');
plt.colorbar();
```
